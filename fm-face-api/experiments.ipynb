```python
from retinaface import RetinaFace
```

```python
faces = RetinaFace.detect_faces("storage/samples/images/sample.jpg")
```

```python
faces
```

```python
faces["face_1"]
```

```python
import cv2
img = cv2.imread("storage/samples/images/sample.jpg")
area = faces["face_2"]["facial_area"]
x1, y1, x2, y2 = area
crop_img = img[y1:y2, x1:x2]
```

```python
from matplotlib import pyplot as plt
```

```python
rgb_img = cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```
